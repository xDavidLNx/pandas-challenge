# PiyCity Schools analysis


```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load
school_data = Path("Resources/schools_complete.csv")
student_data = Path("Resources/students_complete.csv")

# Storing the data in the file in a dataframe
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# Combine the data into a single dataset.  
district_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
district_df.head()
```

```python
# Total number of unique schools based on the school's dataframe
school_count = len(school_data['School ID'].unique())
school_count
```

```python
# Total number of students based on the student dataframe
student_count = student_data['Student ID'].count()
student_count
```

```python
# Calculate the district's total budget for schools
total_budget = school_data['budget'].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = district_df['math_score'].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = district_df['reading_score'].mean()
average_reading_score
```

```python
# % passing math (the percentage of students who passed math)
pass_math = district_df[district_df['math_score']>=70].count()
math_approval_rate = (pass_math['student_name']/student_count)*100
math_approval_rate
```

```python
# % passing reading (the percentage of students who passed reading)
pass_reading = district_df[district_df['reading_score']>=70].count()
reading_approval_rate = (pass_reading['student_name']/student_count)*100
reading_approval_rate
```

```python
#% overall passing (the percentage of students who passed math AND reading) 
overall_pass = district_df[(district_df['math_score']>=70) & (district_df['reading_score']>=70)]
#Second way to resolve it is by applying consecutive filters instead of a filter with two conditions
#overall_pass = district_df[district_df['math_score']>=70]
#overall_pass = overall_pass[overall_pass['reading_score']>=70]
overall_pass_rate = overall_pass['student_name'].count()/student_count*100
overall_pass_rate
```

```python
# High-level snapshot of the district's key metrics
district_summary = pd.DataFrame({
    "Total Students" : [student_count],
    "Total Schools" : [school_count],
    "Total Budget": [total_budget],
    "Average math score": [average_math_score],
    "Average reading score": [average_reading_score],
    "% Passing math":[math_approval_rate],
    "% Passing reading":[reading_approval_rate],
    "Overall passing rate": [overall_pass_rate]

}
)
# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,d}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average math score"] = district_summary["Average math score"].map("{:,.2f}%".format)
district_summary["Average reading score"] = district_summary["Average reading score"].map("{:,.2f}%".format)
district_summary["% Passing math"] = district_summary["% Passing math"].map("{:,.2f}%".format)
district_summary["% Passing reading"] = district_summary["% Passing reading"].map("{:,.2f}%".format)
district_summary["Overall passing rate"] = district_summary["Overall passing rate"].map("{:,.2f}%".format)
# Display the DataFrame
district_summary
```

```python
# Select all of the school types
school_types = school_data.set_index(['school_name'])["type"]
school_types
```

```python
# Calculate the total student count per school
per_school_counts = school_data.groupby('school_name')['size'].sum()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = school_data.groupby('school_name')['budget'].sum()
budget_student = per_school_budget/per_school_counts
budget_student
```

```python
# Calculate the average test scores per school
per_school_math = district_df.groupby('school_name')['math_score'].mean()
per_school_reading = district_df.groupby('school_name')['reading_score'].mean()

avg_scores = pd.merge(per_school_math,per_school_reading, right_index=True,left_index=True)
avg_scores
```

```python
# Calculate the number and proportion of students per school with math scores of 70 or higher
students_passing_math = district_df[district_df['math_score']>=70]
school_students_passing_math = students_passing_math.groupby('school_name')['math_score'].count()
school_students_passing_math_rate = school_students_passing_math/per_school_counts*100
school_students_passing_math_rate
```

```python
# Calculate the number and proportion of students per school with reading scores of 70 or higher
students_passing_reading = district_df[district_df['reading_score']>=70]
school_students_passing_reading = students_passing_reading.groupby('school_name')['reading_score'].count()
school_students_passing_reading_rate = school_students_passing_reading/per_school_counts*100
school_students_passing_reading_rate
```

```python
# % overall passing (the percentage of students who passed math AND reading)
overall_pass_perschool = district_df[(district_df['math_score']>=70) & (district_df['reading_score']>=70)]
overall_pps_count = overall_pass_perschool.groupby('school_name')['reading_score'].count()
overall_pps_count_rate = overall_pps_count/per_school_counts*100
overall_pps_count_rate
```

```python
# High-level snapshot of the district's key metrics
per_school_summary = pd.DataFrame({
    "School Types" : school_types,
    "Total students" : per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": budget_student,
    "Average math score": per_school_math,
    "Average reading score": per_school_reading,
    "% Passing math": school_students_passing_math_rate,
    "% Passing reading": school_students_passing_reading_rate,
    "Overall passing rate": overall_pps_count_rate

}
)

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
#per_school_summary["Average math score"] = per_school_summary["Average math score"].map("{:,.2f}".format)
#per_school_summary["Average reading score"] = per_school_summary["Average reading score"].map("{:,.2f}".format)
#per_school_summary["% Passing math"] = per_school_summary["% Passing math"].map("{:,.2f}%".format)
#per_school_summary["% Passing reading"] = per_school_summary["% Passing reading"].map("{:,.2f}%".format)
#per_school_summary["Overall passing rate"] = per_school_summary["Overall passing rate"].map("{:,.2f}%".format)
# Display the DataFrame
```

```python
# Use the code provided to separate the data by grade
grade9 = district_df[(district_df['grade']=='9th')]
grade10 = district_df[(district_df['grade']=='10th')]
grade11 = district_df[(district_df['grade']=='11th')]
grade12 = district_df[(district_df['grade']=='12th')]

# Group by `school_name` and take the mean of the `math_score` column for each.
grade9_reading = grade9.groupby('school_name')['reading_score'].mean()
grade10_reading = grade10.groupby('school_name')['reading_score'].mean()
grade11_reading = grade11.groupby('school_name')['reading_score'].mean()
grade12_reading = grade12.groupby('school_name')['reading_score'].mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`

avg_grades_read_df = pd.DataFrame({"9th":grade9_reading,
                              "10th":grade10_reading,
                              "11th":grade11_reading,
                              "12th":grade12_reading})
avg_grades_read_df.reset_index()
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels_bins = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per student)"] = pd.cut(budget_student,spending_bins,labels = labels_bins)
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per student)"])["Average math score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per student)"])["Average reading score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per student)"])["% Passing math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per student)"])["% Passing reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per student)"])["Overall passing rate"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({"Average math Score": spending_math_scores,      
                                    "Average reading scores": spending_reading_scores,
                                    "% Passing math": spending_passing_math,
                                    "% Passing reading": spending_passing_reading,
                                    "% Overall Passing rate": overall_passing_spending})

# Display results
spending_summary
```

## Scores by School Size

```python
#Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels_bins = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_counts,size_bins,labels=labels_bins)
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average math score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average reading score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["Overall passing rate"].mean()
```

```python
# Assemble into DataFrame
size_summary = pd.DataFrame({"Average math Score": size_math_scores,      
                                    "Average reading scores": size_reading_scores,
                                    "% Passing math": size_passing_math,
                                    "% Passing reading": size_passing_reading,
                                    "% Overall Passing rate": size_overall_passing})

# Display results
size_summary
```

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
average_math_score_by_type = per_school_summary.groupby(["School Types"])["Average math score"].mean()
average_reading_score_by_type = per_school_summary.groupby(["School Types"])["Average reading score"].mean()
average_percent_passing_math_by_type = per_school_summary.groupby(["School Types"])["% Passing math"].mean()
average_percent_passing_reading_by_type = per_school_summary.groupby(["School Types"])["% Passing reading"].mean()
average_percent_overall_passing_by_type = per_school_summary.groupby(["School Types"])["Overall passing rate"].mean()
```

```python
# Assemble into DataFrame
type_summary = pd.DataFrame({"Average math Score": average_math_score_by_type,      
                                    "Average reading scores": average_reading_score_by_type,
                                    "% Passing math": average_percent_passing_math_by_type,
                                    "% Passing reading": average_percent_passing_reading_by_type,
                                    "% Overall Passing rate": average_percent_overall_passing_by_type})

# Display results
type_summary
```
